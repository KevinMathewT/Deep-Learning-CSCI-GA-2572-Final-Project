```python
import wandb

# Set your project and entity
project_name = "DL Final Project"  # Project name
entity_name = "kevinmathew"        # Entity name

# Authenticate wandb
wandb.login()

# Fetch all runs in the project
api = wandb.Api()
runs = api.runs(f"{entity_name}/{project_name}")

# Sort runs by creation time
sorted_runs = sorted(runs, key=lambda run: run.created_at, reverse=True)

# Retain the latest run and delete others
if len(sorted_runs) > 1:
    for run in sorted_runs[1:]:
        print(f"Deleting run: {run.name} (ID: {run.id})")
        run.delete()  # Deletes the run
else:
    print("Only one run exists, no need to delete others.")
```

```python
import wandb
import pandas as pd
import re

# Initialize W&B API
api = wandb.Api()

# Define your project details
entity = "kevinmathew"
project = "DL Final Project"

# Initialize a list to store run data
run_data = []

# Fetch all runs for the specified project
runs = api.runs(f"{entity}/{project}")

# Regular expression to extract 'wall loss' values
wall_loss_pattern = re.compile(r"wall loss:\s*([+-]?\d+\.?\d*)")

# Iterate over each run
for run in runs:
    # Access the 'output.log' file
    try:
        log_file = run.file("output.log")
        if log_file is None:
            print(f"No log file found for run {run.id}")
            continue
        log_content = log_file.download(replace=True).read()
        if isinstance(log_content, bytes):
            log_content = log_content.decode("utf-8")
    except Exception as e:
        print(f"Error downloading log for run {run.id}: {e}")
        continue


    # Extract all 'wall loss' values from the log content
    wall_losses = [float(match) for match in wall_loss_pattern.findall(log_content)]

    # Determine the best (minimum) 'wall loss' value for the run
    best_wall_loss = min(wall_losses) if wall_losses else None

    # Append the run ID and best wall loss to the data list
    run_data.append({"run_id": run.id, "best_wall_loss": best_wall_loss})

# Create a DataFrame from the collected data
df = pd.DataFrame(run_data)

# Drop rows where 'best_wall_loss' is None
df = df.dropna(subset=["best_wall_loss"])

# Sort the DataFrame by 'best_wall_loss' in ascending order
df = df.sort_values(by="best_wall_loss", ascending=True)

# Reset the index for the sorted DataFrame
df = df.reset_index(drop=True)

# Display the DataFrame
df
```
